```python
import pandas as pd
from textblob import TextBlob
import re
import sentiment_mod as s
```

```python
df = pd.read_csv('Output/recent_tweets.csv')
```

```python
#clean tweet texts
def clean(text):
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('<.*?>+', '', text)
    return text
```

```python
df['normalized_date'] = pd.to_datetime(df['date'],errors='coerce').dt.date
#df['normalized_date'] = datetime(df['date']
df.dropna(subset=['normalized_date','text'],inplace=True)
#dates.dropna(inplace=True)   
```

```python
df["TMP"] = df['normalized_date']                # index is a DateTimeIndex
df = df[df.TMP.notnull()]                  # remove all NaT values
df.drop(["TMP"], axis=1, inplace=True) 
```

```python
def analyze_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean(tweet))
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
df['polarity'] = df['text'].apply(lambda x: TextBlob(clean(x)).sentiment.polarity)
df['subjectivity'] = df['text'].apply(lambda x: TextBlob(clean(x)).sentiment.subjectivity)
#df('TextBlob_SA') = df['text'].apply(lambda x: analyze_sentiment(x))
df['TextBlob_SA'] = [ analyze_sentiment(tweet) for tweet in df['text'] ]
df.head()
```

```python
%%time
SV = df['text'].apply(lambda x: s.sentiment(clean(x))).values
```

```python
nltk_sentiment = []
nltk_confidence = []
[[nltk_sentiment.append(i[0]),nltk_confidence.append(i[1])] for i in SV]
```

```python
df['NLTK_sentiment'] =  nltk_sentiment
df['NLTK_confidence'] = nltk_confidence
```

```python
df['NLTK_SA'] = df['NLTK_sentiment'].map({'pos':1,'neg':-1})
df['NLTK_wSA'] = df['NLTK_SA'] * df['NLTK_confidence']
```

```python
#Vader Sentiment Analysis
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


analyser = SentimentIntensityAnalyzer()
scores=[]
for i in range(len(df['text'])):
    
    score = analyser.polarity_scores(df['text'][i])
    score = score['compound']
    scores.append(score)
    print(i, score)
df['VaderScore']=scores    

sentiment=[]
for i in scores:
    if i>=0.05:
        sentiment.append('Positive')
    elif i<=(-0.05):
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
df['Vader_Sentiment']=pd.Series(sentiment)




df.head()
```

```python
df['Vader_SA'] = df['Vader_Sentiment'].map({'Positive':1,'Negative':-1, 'Neutral':0})
```
